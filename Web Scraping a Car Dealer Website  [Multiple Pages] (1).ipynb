```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D&list_price_max&maximum_distance=all&zip'
```

Requesting

```python
response = requests.get(website)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
results = soup.find_all('div', {'class':'vehicle-card'})
```

# Put everything inside a For- Loop

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for result in results:
    #name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')
        
    #mileage
    try:
        mileage.append(result.find('div', {'class':'mileage'}).get_text())
    except:
        name.append('n/a')
        
     #dealers name
    try:
        dealer_name.append(result.find('strong').get_text().strip())
    except:
        name.append('n/a')
        
     #Rating
    try:
        rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
    except:
        name.append('n/a')
        
     #Rating count
    try:
        rating_count.append(result.find('span', {'class':'sds-rating__link sds-button-link'}).get_text())
    except:
        name.append('n/a')
        
     #Price
    try:
        price.append(results[0].find('span', {'class':'primary-price'}).get_text())
    except:
        name.append('n/a')
```


# Create Pandas DataFrame

```python
car_dealers= pd.DataFrame({'Name':name,'Mileage':mileage, 'Dealer Name':dealer_name,
                          'Rating':rating, 'Rating Count':rating_count, "Price": price})
```

```python
car_dealers
```

# Data Cleaning using pandas

```python
car_dealers['Rating Count'] = car_dealers['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
```

```python
car_dealers
```

```python
car_dealers.to_excel('single_page_car.xlsx', index= False)
```

# Part 2 : Pagination (including multiple pages)

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for i in range(1,15):
   
    #website is a variable
    website = 'https://www.cars.com/shopping/results/?page=' + str(i) +'&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=all&models[]=&stock_type=cpo&zip='
    
    #requesting for website
    response = requests.get(website)
    
    #soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #results object
    results = soup.find_all('div', {'class':'vehicle-card'})
    
    
    
    for result in results:
        #name
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')

        #mileage
        try:
            mileage.append(result.find('div', {'class':'mileage'}).get_text())
        except:
            name.append('n/a')

         #dealers name
        try:
            dealer_name.append(result.find('strong').get_text().strip())
        except:
            name.append('n/a')

         #Rating
        try:
            rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            name.append('n/a')

         #Rating count
        try:
            rating_count.append(result.find('span', {'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            name.append('n/a')

         #Price
        try:
            price.append(results[0].find('span', {'class':'primary-price'}).get_text())
        except:
            name.append('n/a')
```

```python
car_dealers= pd.DataFrame({'Name':name,'Mileage':mileage, 'Dealer Name':dealer_name,
                          'Rating':rating, 'Rating Count':rating_count, "Price": price})
```
